# Capital Asset Pricing Model 
1. It assumes the existence of a risk free asset. An investment with zero risk which in turn means zero standard deviation.
2. It has a positive rate of return but zero risk associated with it. So ppl will prefer to buy this. people will be automatically drawn towards lower rate of return since risk factor is eliminatd over here.
3. However in efficient markets, investors are only compensated for the added risk they are willing to bear. In such settings, higher rate of return means more risk and vice versa. 
4. Risk free rate will provide a low level of expected return.
5. A rational investor would form a portfolio considering both the risk free rate and the market portfolio.
6. How much one wants to invest in risk free which has low rate of returns and market portfolio which has higher rate of return will depend on how much they want to earn.
7. The line that connects the risk free rate and is tangent to the efficient frontier is called the capital market line.
8. Every investor can invest in a combination between the risk free asset and the market portfolio according to his/her risk apetite. If one is able to risk more, they are able to hold a greater portion of market portfolio and in the opposite case if one is unwilling to risk, they will buy more of the risk free asset and less of the market portfolio.

# Beta 
1. One of the main pillars of CAPM. Helps us quantify the relationship between a security and overall market portfolio.
2. If there is a recession, then it is reasonable to expect the prices of most assets that make up the market portfolio will decrease and the market portfolio will experience a negative rate of return.
3. Investors cannot protect themselves through diversification because it is systemic risk.
4. ***Safer Stocks*** - They will earn less than the market portfolio when the economy grows, ***Riskier Stocks*** - Will earn more than the market portfolio when the economy grows.
5. Beta allows us to measure the relationship between a stock and the market portfolio. ***Beta measures the market risk that cannot be avoided through diversification***.
6. Riskier the stock, higher it's beta value. 
7. A beta value of zero = np relationship between a stock and market, beta <1 means the stock is defensive and beta > 1 means the stock is aggressive. Aggressive stock will give better returns when the economy flourishes and lose more when it goes down.  
8. Beta = covariance(stock, market)/(variance of the market)
9. Beta is a measurement that shows us how risky an individual security is regarding the rest of the market.
